```python

def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['Purchase Hour'] = pd.to_datetime(df['Time of Purchase']).dt.hour
df['Time Period'] = df['Purchase Hour'].apply(time_of_day)


refund_analysis = df.groupby(['Time Period', 'Ticket Type'])['Refund Request'].value_counts(normalize=True).unstack().fillna(0)

print(refund_analysis)


refund_analysis.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Refund Requests by Time Period and Ticket Type')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python

df['Date of Purchase'] = pd.to_datetime(df['Date of Purchase'])


if 'Refund Request' in df.columns:
    df = df[df['Refund Request'].str.lower() != 'yes']


df = df[df['Price'] > 0]


df['number_of_rides'] = 1

monthly = (
    df
    .set_index('Date of Purchase')
    .resample('M')
    .agg({'number_of_rides': 'sum', 'Price': 'sum'})
    .reset_index()
)
monthly['month_num'] = np.arange(len(monthly))


next_m = monthly['month_num'].max() + 1


model_r = LinearRegression().fit(
    monthly[['month_num']], monthly['number_of_rides']
)
pred_rides = model_r.predict([[next_m]])[0]

model_p = LinearRegression().fit(
    monthly[['month_num']], monthly['Price']
)
pred_revenue = model_p.predict([[next_m]])[0]


print(f'     : {int(pred_rides)} ')
print(f'    : {pred_revenue:,.2f} ')


last_date = monthly['Date of Purchase'].max() + pd.DateOffset(months=1)
forecast_row = {
    'Date of Purchase': last_date,
    'number_of_rides': pred_rides,
    'Price': pred_revenue,
    'month_num': next_m
}


monthly_plot = pd.concat(
    [monthly, pd.DataFrame([forecast_row])],
    ignore_index=True
)


fig, axes = plt.subplots(1, 2, figsize=(14, 5), constrained_layout=True)


axes[0].plot(
    monthly_plot['Date of Purchase'],
    monthly_plot['number_of_rides'],
    marker='o', linestyle='-',
    label='Actual & Forecast'
)
axes[0].set_title('Forecast of Train Rides')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Number of Rides')
axes[0].tick_params(axis='x', rotation=45)


axes[1].plot(
    monthly_plot['Date of Purchase'],
    monthly_plot['Price'],
    marker='o', linestyle='-',
    color='green',
    label='Actual & Forecast'
)
axes[1].set_title('Forecast of Revenue')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Revenue (EGP)')
axes[1].tick_params(axis='x', rotation=45)

plt.show()
```
